```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import DBSCAN
import pandas
```

```python
data = pandas.read_csv("partisanship_polarization_negativity.csv")
data = data.to_numpy()

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,1], data[:,2], data[:,3], s=80)
ax.set_xlabel('strength of partisan identity')
ax.set_ylabel('affective polarization')
ax.set_zlabel('perceived campaign negativity')
ax.view_init(azim=200)
plt.show()

model = DBSCAN(eps=1, min_samples=10)
model.fit_predict(data[:,1:4])
pred = model.fit_predict(data[:,1:4])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(data[:,1], data[:,2], data[:,3], c=model.labels_, s=80)
ax.set_xlabel('strength of partisan identity')
ax.set_ylabel('affective polarization')
ax.set_zlabel('perceived campaign negativity')
ax.view_init(azim=200)
plt.show()

print("number of cluster identified: {}".format(len(set(model.labels_))))
```
